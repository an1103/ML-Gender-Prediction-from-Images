```python
female_path = glob('./data/female/*.jpg')
male_path = glob('./data/male/*.jpg')
```

```python
#considering one image
path1 = female_path[5]
img = cv2.imread(path1)
plt.imshow(img)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap = 'gray')
```

```python
haar = cv2.CascadeClassifier('./Model/haarcascade_frontalface_default.xml')
face = haar.detectMultiScale(gray, 1.05, 5)
print(face)
```

```python
for x, y, w, h in face:
    cv2.rectangle(img, (x, y), (x+w, y+h), (0, 0, 255), 2)
plt.imshow(img)
```

```python
crop_img = img[y:y+h, x:x+w]
plt.imshow(crop_img)
```

```python
cv2.imwrite('f_01.png', crop_img)
```

```python
#applying to all images
def crop_all_images(path, gender, i):
    img = cv2.imread(path)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    #get face dimensions
    haar = cv2.CascadeClassifier('./Model/haarcascade_frontalface_default.xml')
    face = haar.detectMultiScale(gray, 1.05, 5)
    
    for x, y, w, h in face:
        cv2.rectangle(img, (x, y), (x+w, y+h), (0, 0, 255), 2)
    crop_img = img[y:y+h, x:x+w]
    if gender == 'male':
        cv2.imwrite('./data/crop/male_crop/{}_{}.png'.format(gender, i), crop_img)
    else:
        cv2.imwrite('./data/crop/female_crop/{}_{}.png'.format(gender, i), crop_img)
    
    
#applying function
for i, path in enumerate(female_path):
    try:
        crop_all_images(path, 'female', i)
    except:
        print('image {} cannot be processed'.format(i))

for i, path in enumerate(male_path):
    try:
        crop_all_images(path, 'male', i)
    except:
        print('image {} cannot be processed'.format(i))
```
